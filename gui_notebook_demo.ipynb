## Demo: Using the Electric Field Solver

Launching the electric field solver window is very simple.

Via the Jupyter Notebook environment, we simply import the `solver.py` module, then launch using the `create()` command.
Alternatively, `solver.py` can be run as a stand-alone Python script.

This opens up the user interface window, where the initial conditions can be set using a variety of shapes, colours, and potential values set directly for the boundary conditions.

![alt](GUI_screenshots/gui_sceenshot_1.png)

```python
import solver
```

```python
#launch the electric field solver window using the "create()" command
g = solver.create()
```

```python
#create gallery of plot types
g.plot("initial")  # the input potentials the user set on the GUI canvas
g.plot("mask")  # the Boolean mask showing whether points are boundaries or background
g.plot("solution")  # the solved potential values from the processed canvas
g.plot("E_x")  # x-components of the calculated electric field
g.plot("E_y")  # y-components of the calculated electric field
g.plot("field")  # output plot of the solved field lines
```
